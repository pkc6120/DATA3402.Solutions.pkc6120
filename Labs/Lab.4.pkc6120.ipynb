1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.value = 0
        self.max_value = max_value

    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("Error: Counter reached max value")

    def reset(self):
        self.value = 0

    def get_value(self):
        return self.value

max_value = 5
counter = Counter(max_value)

for _ in range(max_value + 2):
    counter.increment()
    print("Current Value:", counter.get_value())

print("Reset Count")
counter.reset()
print("Current Value Now:", counter.get_value())
```

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        raise NotImplementedError("Error: No area method")

    def perimeter(self):
        raise NotImplementedError("Error: No perimeter method")
    
    def point_check(self, x, y):
        raise NotImplementedError("Error: No point check method")


class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def area(self):
        return self.get_length() * self.get_width()

    def perimeter(self):
        return 2 * (self.get_length() + self.get_width())

    def point_check(self, x, y):
        return self.get_x() <= x <= (self.get_x() + self.get_length()) and \
               self.get_y() <= y <= (self.get_y() + self.get_width())


class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius

    def get_radius(self):
        return self.__radius

    def area(self):
        return math.pi * self.get_radius() ** 2

    def perimeter(self):
        return 2 * math.pi * self.get_radius()

    def point_check(self, x, y):
        distance = math.sqrt((x - self.get_x()) ** 2 + (y - self.get_y()) ** 2)
        return distance <= self.get_radius()


class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        super().__init__(x, y)
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def get_side1(self):
        return self.__side1

    def get_side2(self):
        return self.__side2

    def get_side3(self):
        return self.__side3

    def area(self):
        s = (self.__side1 + self.__side2 + self.__side3) / 2
        area = math.sqrt(s * (s - self.__side1) * (s - self.__side2) * (s - self.__side3))
        return area

    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3

    def point_check(self, x, y):
        detT = (self.get_x() - x) * (self.get_y() + self.get_side1() - y) - \
               (self.get_x() + self.get_side2() - x) * (self.get_y() - y)
        detA = (self.get_x() - self.get_x() - x) * (self.get_y() + self.get_side1() - self.get_y()) - \
               (self.get_x() + self.get_side2() - self.get_x()) * (self.get_y() - self.get_y())

        detB = (self.get_x() - x) * (self.get_y() - self.get_y()) - \
               (self.get_x() - self.get_x()) * (self.get_y() - y)

        u = detT / detA
        v = detB / detA

        return u >= 0 and v >= 0 and u + v <= 1


#Example
rectangle = Rectangle(5, 3, 2, 1)
circle = Circle(4, 2, 3)
triangle = Triangle(3, 4, 5, 1, 2)

print("Is (3, 2) inside the Rectangle?", rectangle.point_check(3, 2))
print("Is (2, 4) inside the Circle?", circle.point_check(2, 4))
print("Is (2, 3) inside the Triangle?", triangle.point_check(2, 3))
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        raise NotImplementedError("Error: No area method")

    def perimeter(self):
        raise NotImplementedError("Error: No perimeter method")
    
    def point_check(self, x, y):
        raise NotImplementedError("Error: No point check method")
    
    def bound_points(self):
        raise NotImplementedError("Error: No boundary point method")
        
    def overlap(self, other_shape):
        raise NotImplementedError("Error: No overlap method")
        
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def area(self):
        return self.get_length() * self.get_width()

    def perimeter(self):
        return 2 * (self.get_length() + self.get_width())

    def point_check(self, x, y):
        return self.get_x() <= x <= (self.get_x() + self.get_length()) and \
               self.get_y() <= y <= (self.get_y() + self.get_width())
    
    def bound_points(self):
        points = []
        for i in range(16):
            if i < self.get_length():
                points.append((self.get_x() + i, self.get_y()))
            elif i < self.get_length() + self.get_width():
                points.append((self.get_x() + self.get_length(), self.get_y() + i - self.get_length()))
            else:
                points.append((self.get_x() + self.get_length() - (i - self.get_length() - self.get_width()), 
                               self.get_y() + self.get_width()))
        return points
    
    def overlap(self, other_shape):
        if isinstance(other_shape, Rectangle):
            return not (self.get_x() + self.get_length() < other_shape.get_x() or
                        self.get_x() > other_shape.get_x() + other_shape.get_length() or
                        self.get_y() + self.get_width() < other_shape.get_y() or
                        self.get_y() > other_shape.get_y() + other_shape.get_width())
        elif isinstance(other_shape, Circle):
            closest_x = max(self.get_x(), min(other_shape.get_x(), self.get_x() + self.get_length()))
            closest_y = max(self.get_y(), min(other_shape.get_y(), self.get_y() + self.get_width()))
            distance_squared = (closest_x - other_shape.get_x()) ** 2 + (closest_y - other_shape.get_y()) ** 2
            return distance_squared < other_shape.get_radius() ** 2
        elif isinstance(other_shape, Triangle):
            return False
        else:
            raise NotImplementedError("Overlap detection not implemented for the provided shape.")


class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius

    def get_radius(self):
        return self.__radius

    def area(self):
        return math.pi * self.get_radius() ** 2

    def perimeter(self):
        return 2 * math.pi * self.get_radius()

    def point_check(self, x, y):
        distance = math.sqrt((x - self.get_x()) ** 2 + (y - self.get_y()) ** 2)
        return distance <= self.get_radius()
    
    def bound_points(self):
        points = []
        for i in range(16):
            angle = 2 * math.pi * i / 16
            x_point = self.get_x() + self.get_radius() * math.cos(angle)
            y_point = self.get_y() + self.get_radius() * math.sin(angle)
            points.append((x_point, y_point))
        return points
    
    def overlap(self, other_shape):
        if isinstance(other_shape, Circle):
            distance_between_centers = math.sqrt((self.get_x() - other_shape.get_x()) ** 2 +
                                                (self.get_y() - other_shape.get_y()) ** 2)
            return distance_between_centers < self.get_radius() + other_shape.get_radius()
        elif isinstance(other_shape, Rectangle):
            return other_shape.overlap(self)
        elif isinstance(other_shape, Triangle):
            return False
        else:
            raise NotImplementedError("Overlap detection not implemented for the provided shape.")


class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        super().__init__(x, y)
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def get_side1(self):
        return self.__side1

    def get_side2(self):
        return self.__side2

    def get_side3(self):
        return self.__side3

    def area(self):
        s = (self.__side1 + self.__side2 + self.__side3) / 2
        area = math.sqrt(s * (s - self.__side1) * (s - self.__side2) * (s - self.__side3))
        return area

    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3

    def point_check(self, x, y):
        detT = (self.get_x() - x) * (self.get_y() + self.get_side1() - y) - \
               (self.get_x() + self.get_side2() - x) * (self.get_y() - y)
        detA = (self.get_x() - self.get_x() - x) * (self.get_y() + self.get_side1() - self.get_y()) - \
               (self.get_x() + self.get_side2() - self.get_x()) * (self.get_y() - self.get_y())

        detB = (self.get_x() - x) * (self.get_y() - self.get_y()) - \
               (self.get_x() - self.get_x()) * (self.get_y() - y)

        u = detT / detA
        v = detB / detA

        return u >= 0 and v >= 0 and u + v <= 1
    
    def bound_points(self):
        points = []
        for i in range(16):
            if i < self.get_side1():
                x_point = self.get_x() + i
                y_point = self.get_y()
            elif i < self.get_side1() + self.get_side2():
                x_point = self.get_x() + self.get_side1()
                y_point = self.get_y() + i - self.get_side1()
            else:
                x_point = self.get_x() + self.get_side1() - (i - self.get_side1() - self.get_side2())
                y_point = self.get_y() + self.get_side2()
            points.append((x_point, y_point))
        return points
    
    def overlap(self, other_shape):
        if isinstance(other_shape, Triangle):
            return False
        elif isinstance(other_shape, (Rectangle, Circle)):
            return other_shape.overlap(self)
        else:
            raise NotImplementedError("Overlap detection not implemented for the provided shape.")
            
            

    
#Example
rectangle1 = Rectangle(5, 3, 2, 1)
rectangle2 = Rectangle(4, 2, 3, 2)
circle = Circle(4, 6, 6)
triangle = Triangle(3, 4, 5, 1, 2)

print("Rectangle 1 overlaps with Rectangle 2:", rectangle1.overlap(rectangle2))
print("Rectangle overlaps with Circle:", rectangle1.overlap(circle))
print("Triangle overlaps with Rectangle:", triangle.overlap(rectangle1))
```

```python
class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        raise NotImplementedError("Error: No area method")

    def perimeter(self):
        raise NotImplementedError("Error: No perimeter method")
    
    def point_check(self, x, y):
        raise NotImplementedError("Error: No point check method")
    
    def bound_points(self):
        raise NotImplementedError("Error: No boundary point method")
        
    def overlap(self, other_shape):
        raise NotImplementedError("Error: No overlap method")
        
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def area(self):
        return self.get_length() * self.get_width()

    def perimeter(self):
        return 2 * (self.get_length() + self.get_width())

    def point_check(self, x, y):
        return self.get_x() <= x <= (self.get_x() + self.get_length()) and \
               self.get_y() <= y <= (self.get_y() + self.get_width())
    
    def bound_points(self):
        points = []
        for i in range(16):
            if i < self.get_length():
                points.append((self.get_x() + i, self.get_y()))
            elif i < self.get_length() + self.get_width():
                points.append((self.get_x() + self.get_length(), self.get_y() + i - self.get_length()))
            else:
                points.append((self.get_x() + self.get_length() - (i - self.get_length() - self.get_width()), 
                               self.get_y() + self.get_width()))
        return points
    
    def overlap(self, other_shape):
        if isinstance(other_shape, Rectangle):
            return not (self.get_x() + self.get_length() < other_shape.get_x() or
                        self.get_x() > other_shape.get_x() + other_shape.get_length() or
                        self.get_y() + self.get_width() < other_shape.get_y() or
                        self.get_y() > other_shape.get_y() + other_shape.get_width())
        elif isinstance(other_shape, Circle):
            closest_x = max(self.get_x(), min(other_shape.get_x(), self.get_x() + self.get_length()))
            closest_y = max(self.get_y(), min(other_shape.get_y(), self.get_y() + self.get_width()))
            distance_squared = (closest_x - other_shape.get_x()) ** 2 + (closest_y - other_shape.get_y()) ** 2
            return distance_squared < other_shape.get_radius() ** 2
        elif isinstance(other_shape, Triangle):
            return False
        else:
            raise NotImplementedError("Overlap detection not implemented for the provided shape.")


class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius

    def get_radius(self):
        return self.__radius

    def area(self):
        return math.pi * self.get_radius() ** 2

    def perimeter(self):
        return 2 * math.pi * self.get_radius()

    def point_check(self, x, y):
        distance = math.sqrt((x - self.get_x()) ** 2 + (y - self.get_y()) ** 2)
        return distance <= self.get_radius()
    
    def bound_points(self):
        points = []
        for i in range(16):
            angle = 2 * math.pi * i / 16
            x_point = self.get_x() + self.get_radius() * math.cos(angle)
            y_point = self.get_y() + self.get_radius() * math.sin(angle)
            points.append((x_point, y_point))
        return points
    
    def overlap(self, other_shape):
        if isinstance(other_shape, Circle):
            distance_between_centers = math.sqrt((self.get_x() - other_shape.get_x()) ** 2 +
                                                (self.get_y() - other_shape.get_y()) ** 2)
            return distance_between_centers < self.get_radius() + other_shape.get_radius()
        elif isinstance(other_shape, Rectangle):
            return other_shape.overlap(self)
        elif isinstance(other_shape, Triangle):
            return False
        else:
            raise NotImplementedError("Overlap detection not implemented for the provided shape.")


class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        super().__init__(x, y)
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def get_side1(self):
        return self.__side1

    def get_side2(self):
        return self.__side2

    def get_side3(self):
        return self.__side3

    def area(self):
        s = (self.__side1 + self.__side2 + self.__side3) / 2
        area = math.sqrt(s * (s - self.__side1) * (s - self.__side2) * (s - self.__side3))
        return area

    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3

    def point_check(self, x, y):
        detT = (self.get_x() - x) * (self.get_y() + self.get_side1() - y) - \
               (self.get_x() + self.get_side2() - x) * (self.get_y() - y)
        detA = (self.get_x() - self.get_x() - x) * (self.get_y() + self.get_side1() - self.get_y()) - \
               (self.get_x() + self.get_side2() - self.get_x()) * (self.get_y() - self.get_y())

        detB = (self.get_x() - x) * (self.get_y() - self.get_y()) - \
               (self.get_x() - self.get_x()) * (self.get_y() - y)

        u = detT / detA
        v = detB / detA

        return u >= 0 and v >= 0 and u + v <= 1
    
    def bound_points(self):
        points = []
        for i in range(16):
            if i < self.get_side1():
                x_point = self.get_x() + i
                y_point = self.get_y()
            elif i < self.get_side1() + self.get_side2():
                x_point = self.get_x() + self.get_side1()
                y_point = self.get_y() + i - self.get_side1()
            else:
                x_point = self.get_x() + self.get_side1() - (i - self.get_side1() - self.get_side2())
                y_point = self.get_y() + self.get_side2()
            points.append((x_point, y_point))
        return points
    
    def overlap(self, other_shape):
        if isinstance(other_shape, Triangle):
            return False
        elif isinstance(other_shape, (Rectangle, Circle)):
            return other_shape.overlap(self)
        else:
            raise NotImplementedError("Overlap detection not implemented for the provided shape.")


class CompoundShape(Shape):
    def __init__(self, shapes, x, y):
        super().__init__(x, y)
        self.__shapes = shapes

    def get_shapes(self):
        return self.__shapes

    def area(self):
        total_area = 0
        for shape in self.__shapes:
            total_area += shape.compute_area()
        return total_area

    def perimeter(self):
        total_perimeter = 0
        for shape in self.__shapes:
            total_perimeter += shape.compute_perimeter()
        return total_perimeter

    def overlap(self, other_shape):
        for shape in self.__shapes:
            if shape.overlap(other_shape):
                return True
        return False

    def point_check(self, x, y):
        for shape in self.__shapes:
            if shape.is_point_inside(x, y):
                return True
        return False

    def bound_points(self):
        boundary_points = []
        for shape in self.__shapes:
            boundary_points.extend(shape.bound_points())
        return boundary_points

            
            
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
        
        
# Example
if __name__ == "__main__":
    canvas = Canvas(20, 20)

    rectangle = Rectangle(5, 3, 2, 1)
    circle = Circle(4, 10, 10)
    triangle = Triangle(3, 4, 5, 1, 2)

    compound_shape = CompoundShape([rectangle, circle, triangle], 0, 0)

    canvas.clear_canvas()

    # Draw shapes on the canvas
    for x, y in compound_shape.bound_points():
        canvas.set_pixel(int(x), int(y))

    # Display the canvas
    canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def paint(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            for x, y in shape.bound_points():
                if 0 <= x < self.width and 0 <= y < self.height:
                    self.canvas.set_pixel(int(x), int(y))

    def modify_shape(self, index, new_shape):
        if 0 <= index < len(self.shapes):
            self.shapes[index] = new_shape
        else:
            print("Invalid shape index.")

#Example
if __name__ == "__main__":
    drawing = RasterDrawing(20, 20)

    rectangle = Rectangle(5, 3, 2, 1)
    circle = Circle(4, 10, 10)
    triangle = Triangle(3, 4, 5, 1, 2)

    drawing.add_shape(rectangle)
    drawing.add_shape(circle)
    drawing.add_shape(triangle)

    print("Original Drawing:")
    drawing.paint()
    drawing.canvas.display()

    modified_rectangle = Rectangle(3, 6, 2, 1)
    drawing.modify_shape(0, modified_rectangle)

    print("\nDrawing after Modifying Rectangle:")
    drawing.paint()
    drawing.canvas.display()

    modified_circle = Circle(5, 14, 10)
    drawing.modify_shape(1, modified_circle)

    print("\nDrawing after Modifying Circle:")
    drawing.paint()
    drawing.canvas.display()
```
